```python
def f(x):
    return(x ** 4 + 7 * x ** 3 + 5 * x ** 2 - 17 * x + 3)
#Gradient of f
def g(x):
    return(4 * x ** 3 + 21 * x ** 2 + 10 * x - 17)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-6,2,100)
y = f(x)
plt.plot(x,y)
plt.show()
```

```python
x = -2
gamma = 0.01
for i in range(20):
  x = x - gamma * g(x)
  print(x)
```

```python
#Random initialization of x
import numpy as np
np.random.seed(123)
x = np.random.uniform(-6, 2, 5)
x
```

```python
#Gradient Descent Function
#x_i+1 = x_i - gamma * g(x_i)
def grad_descent(x, gamma = 0.01, iter = 3):
    all_x = [x]
    for i in range(iter):
        x = x - gamma * g(x)
        all_x.append(x)
    return(all_x[-1])
```

```python
res = grad_descent(x, gamma = 0.01, iter = 100)
```

```python
#Plot f(x) and local minima
import matplotlib.pyplot as plt
x = np.arange(-6, 2, 0.01)
y = f(x)
plt.plot(x, y, color = 'red')
plt.plot(res[0], f(res[0]), color = 'blue', marker = 'o') #frist minimum
plt.plot(res[1], f(res[1]), color = 'blue', marker = 'o') #second minimum
plt.grid()
```

```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
x = np.linspace(-1,1,100)
y = np.linspace(-1,1,100)
X, Y = np.meshgrid(x,y)
Z = X**2 + Y**2
ax.plot_surface(X, Y, Z, cmap=plt.cm.YlGnBu_r)
ax.set_zlim(0, 2)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z=x^2+y^2$')
plt.show()
```
